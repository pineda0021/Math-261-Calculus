## **Python Code: ε–δ Visualization**

### 🐍 Python Simulation: Exploring the ε–δ Definition of a Limit

This simulation allows you to **visualize and interactively explore** the **ε–δ definition of a limit** for a given function, such as:

$
\lim_\limits{x \to a} f(x) = L
$

---

### 🔍 What the Visualization Shows:

- 📈 **Plots the function** $f(x)$
- 🔴 **Epsilon band**: Horizontal lines at \( $ L \pm \varepsilon $\)
- 🔵 **Delta band**: Vertical lines at \( $a \pm \delta$\)
- ✅ **Verifies** that:

$
|f(x) - L| < \varepsilon \quad \text{whenever} \quad 0 < |x - a| < \delta
\$

This helps illustrate the formal definition of a limit and how the function values get arbitrarily close to $ L $ as $ x $ approaches $a$.


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Example function and limit point
def f(x):
    return x**2

a = 2                # x-value approaching
L = f(a)             # expected limit f(2) = 4

# Visualization function
def visualize_epsilon_delta(epsilon=0.5, delta=0.5):
    x = np.linspace(a - 2, a + 2, 1000)
    y = f(x)

    fig, ax = plt.subplots(figsize=(8, 5))

    # Plot function
    ax.plot(x, y, label=r'$f(x)$')
    ax.axvline(a, color='black', linestyle='--', label=r'$x = a$')
    ax.axhline(L, color='green', linestyle='--', label=r'$L = \lim_{x \to a} f(x)$')

    # Epsilon band (horizontal)
    ax.axhline(L + epsilon, color='red', linestyle='--', label=r'$\varepsilon$ band')
    ax.axhline(L - epsilon, color='red', linestyle='--')

    # Delta band (vertical)
    ax.axvline(a + delta, color='blue', linestyle='--', label=r'$\delta$ band')
    ax.axvline(a - delta, color='blue', linestyle='--')

    # Highlight the valid region
    x_valid = x[(np.abs(x - a) < delta) & (x != a)]
    y_valid = f(x_valid)
    ax.plot(x_valid, y_valid, 'o', color='purple', label=r'$|f(x) - L| < \varepsilon$')

    ax.set_title(r'$\varepsilon$-$\delta$ Visualization for $f(x) = x^2$ as $x \to 2$')
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.legend()
    ax.grid(True)
    plt.show()

# Interactive widget
interact(
    visualize_epsilon_delta,
    epsilon=FloatSlider(value=0.5, min=0.01, max=2, step=0.01, description='ε'),
    delta=FloatSlider(value=0.5, min=0.01, max=2, step=0.01, description='δ')
)
```

### 🧠 Interpretation

Adjust **$\varepsilon$** **(epsilon)** and **$\delta$** **(delta)** to see how tightly the function hugs the limit.

The **purple dots** confirm where:

$
|f(x) - L| < \varepsilon \quad \text{whenever} \quad |x - a| < \delta
$

This visual demonstrates the core idea behind the **ε–δ definition of a limit**.
